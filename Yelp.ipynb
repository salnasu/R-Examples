```r
library(readxl)
library(writexl)
library(tidyverse)
```

```r
# count words
d <- read_xlsx("Data/tacos.xlsx", sheet ="Reviews")
d$Review <- str_to_lower(d$Review)
d$Food <-  str_count(d$Review, c("food"))
d$Good <-  str_count(d$Review, c("good"))
```

```r
d
```

```r
# import data
d <- read_xlsx("Data/tacos.xlsx", sheet="Data")
d <- d %>% rename(rest=Restaurant, rating=Rating)
```

```r
d
```

```r
# descriptive
hist(d$rating, breaks = seq(0,5))
table(d$rating)
d %>% group_by(rest) %>% summarise(rating=mean(rating))
```

```r
# histogram rating by restaurant
histogram <- d %>% 
  group_by(rest) %>% 
  summarise(histogram=list(hist(rating, plot=TRUE, main=rest[1])))
```

```r
# correlation
round(cor(d[,-1]),2)
```

```r
# competitor profile: standard method
arriba <-  filter(d, rest=="Arriba Arriba Sunnyside")
cor(arriba[,-1])
```

```r
# competitor profile: advanced method
profile <- d %>% 
  split(.$rest) %>% 
  map(select, -c(rest)) %>% 
  map(cor) # correlation split by restaurant
```

```r
profile
```

```r
profile_df <- round(as.data.frame(profile),3)
profile_df[c(1,40,79,118,157)] # select rating only
write.csv(profile_df, "cors.csv")
```
